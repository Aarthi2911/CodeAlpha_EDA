```python
df["estimated_revenue"] = df["price"] * df["total_sales"]

df[
    [
        "product_name",
        "price",
        "total_sales",
        "estimated_revenue"
    ]
].head()
```

```python
highest_rated = df.sort_values(
    by=["review_score", "review_count"],
    ascending=[False, False]
)[
    [
        "product_name",
        "category",
        "review_score",
        "review_count"
    ]
].head(10)

highest_rated
```

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

price_outliers = df[
    (df["price"] < lower_bound) |
    (df["price"] > upper_bound)
]

print("Number of price outliers:", len(price_outliers))
```

```python
Q1 = df["total_sales"].quantile(0.25)
Q3 = df["total_sales"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

sales_outliers = df[
    (df["total_sales"] < lower_bound) |
    (df["total_sales"] > upper_bound)
]

print("Number of total sales outliers:", len(sales_outliers))
```

```python
correlation_columns = [
    "price",
    "review_score",
    "review_count",
    "total_sales",
    "estimated_revenue"
]

correlation_matrix = df[correlation_columns].corr()

correlation_matrix
```

```python
print("===== KEY INSIGHTS DATA =====")

# 1. Dataset information
print("\n1. DATASET OVERVIEW")
print("Total products:", df["product_id"].nunique())
print("Total categories:", df["category"].nunique())

# 2. Best-selling category
category_sales = df.groupby("category")["total_sales"].sum().sort_values(ascending=False)

print("\n2. CATEGORY PERFORMANCE")
print("Best-selling category:", category_sales.idxmax())
print("Total units sold:", category_sales.max())
print("Lowest-selling category:", category_sales.idxmin())
print("Total units sold:", category_sales.min())

# 3. Top-selling product
top_product = df.loc[df["total_sales"].idxmax()]

print("\n3. TOP-SELLING PRODUCT")
print("Product:", top_product["product_name"])
print("Category:", top_product["category"])
print("Total units sold:", top_product["total_sales"])

# 4. Monthly sales performance
monthly_sales = df[sales_columns].sum()
monthly_sales.index = [f"Month {i}" for i in range(1, 13)]

print("\n4. MONTHLY SALES")
print("Highest sales month:", monthly_sales.idxmax())
print("Units sold:", monthly_sales.max())
print("Lowest sales month:", monthly_sales.idxmin())
print("Units sold:", monthly_sales.min())

# 5. Correlations
price_sales_corr = df["price"].corr(df["total_sales"])
rating_sales_corr = df["review_score"].corr(df["total_sales"])
review_count_sales_corr = df["review_count"].corr(df["total_sales"])

print("\n5. CORRELATIONS")
print("Price vs Total Sales:", round(price_sales_corr, 3))
print("Review Score vs Total Sales:", round(rating_sales_corr, 3))
print("Review Count vs Total Sales:", round(review_count_sales_corr, 3))

# 6. Outliers
print("\n6. OUTLIERS")
print("Number of price outliers:", len(price_outliers))
print("Number of sales outliers:", len(sales_outliers))

# 7. Estimated revenue
print("\n7. ESTIMATED REVENUE")
print("Total estimated annual revenue: $", round(df["estimated_revenue"].sum(), 2))
```

## Key Insights

1. **Dataset Overview:** The dataset contains 1,000 unique e-commerce products distributed across 7 different product categories.

2. **Category Performance:** Books was the best-selling product category with a total of 938,229 units sold, while Home & Kitchen recorded the lowest sales with 742,141 units sold.

3. **Top-Selling Product:** Product_224 from the Electronics category was the highest-selling individual product, recording 9,151 units sold across the 12-month period.

4. **Monthly Sales Trends:** Month 10 recorded the highest overall sales with 514,798 units sold, whereas Month 5 had the lowest sales with 487,194 units sold.

5. **Price and Sales Relationship:** The correlation between product price and total sales was -0.016, indicating virtually no linear relationship between product price and sales volume in this dataset.

6. **Outlier Detection:** The IQR method identified 13 potential outliers in total product sales, highlighting products with unusually high or low sales compared with the overall distribution.

7. **Estimated Revenue:** The total estimated annual revenue across all products was approximately $1.49 billion, calculated by multiplying each product's listed price by its total annual unit sales.

## Conclusion

The exploratory data analysis revealed meaningful patterns in product performance, category-level sales, monthly trends, pricing relationships, and potential anomalies. Books emerged as the strongest category by unit sales, while Product_224 was the highest-selling individual product. Sales remained relatively consistent across the 12 months, with Month 10 recording the highest performance. The near-zero correlation between price and total sales suggests that product price alone was not a significant linear predictor of sales volume in this dataset.
